Algoritmo K-Means

```python
# Importação das bibliotecas
# Criar a matriz de confusão 
from sklearn import metrics
# Criação de gráficos
import matplotlib.pyplot as plt
# Importa o K-Means
from sklearn.cluster import KMeans
# Importa a base de dados
from sklearn import datasets
```

```python
# Carregamento da base de dados Iris, sendo que X terá os atributos e y a
database = datasets.load_iris()
X,y = database.data, database.target
```

```python
# REaliza o agrupamento com o k-means, utilizando 3 clusters para o valor de k, uma vez que a base possui 3 tipos de classe (Setosa, Versicolor, Virgica)
kmeans = KMeans(n_clusters=3).fit(X)
```

```python
# Obtêm os grupos aos quais cada registro da base de dados foi associado
labels = kmeans.labels_
```

```python
# Cria a matriz de confusão 
confusion_matrix = metrics.confusion_matrix(y, labels)
confusion_matrix
```

```python
# Cria dois gráficos comparando os clusters da base de dados e os obtidos pelo KMeans
plt.figure(figsize = ((15, 6)))
plt.subplot(1, 2, 1)
plt.scatter(X[:,0], X[:,1], c = y)
plt.title("Clusters da base", fontsize = 16)
plt.subplot(1, 2, 2)
plt.scatter(X[:,0], X[:,1], c = labels)
plt.title("Clusters KMeans", fontsize = 16)
plt.show()
```
